```
import nbagrapher
```

# Scatter Plot

```
nbagrapher.plot_player_season(players=['iversal01','bryanko01'], stat="points")
```

```
nbagrapher.player_season(players=['iversal01','bryanko01'],stat="PTS",stat_type="PER_POSS",players_label=["Allen Iverson","Kobe Bryant"])
```

```
nbagrapher.plot_team_season(teams=["PHI","GSW"],stat="points",start_season=2013,end_season=2017)
```

# Shot Charts


```
nbagrapher.player_shot_chart_season(player="leonaka01",season=2019,playoffs=True)
```

```
nbagrapher.player_shot_chart_season(player="leonaka01",season=2019,playoffs=True,fig_type="hex",hex_C="makes",hex_minshots=5)
```

```
nbagrapher.player_shot_chart_season(player="leonaka01",season=2019,playoffs=True,fig_type="hex",hex_C="fg%_adj",hex_minshots=5)
```

# Radar Chart

```
nbagrapher.player_season_radar(players=["antetgi01","embiijo01"],season=2020,player_labels=["Giannis","Embiid"],template="shooting")
```

```
nbagrapher.player_season_radar(players=["antetgi01","embiijo01"],season=2020,player_labels=["Giannis","Embiid"],stats=["PF","ORtg","DRtg","FGA","BLK","TOV"],stat_type="PER_POSS")
```
